```python
import pandas as pd
from prophet import Prophet
```

```python
df = pd.read_csv('/Users/souvikpatra/Downloads/ML_prediction_testing/AD_spotifytest.csv - Sheet1 (1).csv')
df
```

```python
df['day'] =  df['partner_report_date'].apply(lambda x: str(x)[:-6])
df['Month'] = df['partner_report_date'].apply(lambda x: str(x)[-5:-3])
df['Year'] = df['partner_report_date'].apply(lambda x: str(x)[-2:])
df['ds'] = pd.DatetimeIndex(df['partner_report_date'])
df.tail()
```

```python

#taken the user input and setting train model for particular
import datetime

date_user=str(input())
format='%Y/%m/%d' 
date_user= datetime.datetime.strptime(date_user, format)
  
# get the date from the datetime using date() 
# function
print(date_user.date())
df=df.loc[(df['ds']<=date_user),['streams','ds']]
display(df)


```
