```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
    
url="https://www.facebook.com/watch/?v=4006442879381111&ref=sharing"
r=requests.get(url).text
soup=BeautifulSoup(r, 'html.parser')
print(soup)
```

```python
mp1=soup.head
mp1
```

```python
text1=str(mp1.find('script',attrs={"nonce":"g3dmPQzS"}))
text2 = text1.replace(text1[0:52],"");
text3 = text2.replace(text1[-9:],"");
print(text3)
```

```python
import json
json_object = json.loads(text3)
print(type(json_object))
```

```python
headline=json_object['headline']
headline
```

```python
Type=json_object['@type']
Type
```
