```python
import pandas as pd
from fastavro import *
from fastavro import parse_schema
from fastavro import writer
from fastavro import reader
```

```python
df=pd.read_csv("/Users/hariharanthirumurugan/Downloads/mov.csv")
```

```python
schema={"name": "avro file","namespace": "data","type": "record",
        "fields":[{"name":"movieid","type": "int"},
                  {"name":"title","type": "string"},
                  {"name":"casts","type":"string"},
                  {"name":"productionid","type":"string"}]}
```

```python
parsed_schema = parse_schema(schema)
```

```python
records = df.to_dict('records')
```

```python
with open('movies_data.avro', 'wb') as out:
    writer(out, parsed_schema, records)
```

```python
avro_records = []
with open('movies_data.avro', 'rb') as fo:
    avro_reader = reader(fo)
    for record in avro_reader:
        avro_records.append(record)
```
